```python
import pandas as pd
final_year = pd.read_csv("final_tmdb_data_2001.csv.gz")
final_year.head()
```

```python
first_year = pd.read_csv("final_tmdb_data_2000.csv.gz")
first_year.head()
```

```python
years_db=pd.merge(final_year, first_year, how='left')
years_db
```

```python
years_db.drop(columns=['belongs_to_collection','homepage'],inplace=True)
```

How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
years_db.dropna(subset=['revenue','budget'],inplace=True)
```

```python
fil_year=years_db[ (years_db['budget'] == 0) & (years_db['revenue'] ==0)].index
years_db.drop(fil_year, inplace = True)
  
years_db
```

- How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
mov_cert= years_db.groupby('certification')
mov_cert.count()
```

```python
import seaborn as sns
sns.countplot(data=years_db, y='certification',order=years_db.certification.value_counts().index);
print(years_db['certification'].value_counts())
```

What is the average revenue per certification category? and
What is the average budget per certification category?

```python
mov_cert= years_db.groupby('certification')
mov_cert.mean()
```
